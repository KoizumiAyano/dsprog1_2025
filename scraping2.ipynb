# 練習問題

```python
import requests
from bs4 import BeautifulSoup

# アクセス先URL
url = 'https://abehiroshi.la.coocan.jp/movie/eiga.htm'


```

```python
# Webページ取得
res = requests.get(url)
res.encoding = 'shift_jis'  # ← 文字化け防止

# BeautifulSoupで解析
soup = BeautifulSoup(res.text, 'html.parser')

# タイトル（映画名）は <b> タグ内にある
titles = [b_tag.get_text(strip=True) for b_tag in soup.find_all('h2')]

# 結果を出力
for t in titles:
    print(t)
```
